#  Иерархическая кластеризация с алгоритмом объединения single-link clustering

Особенностью этого метода, является то, что он разбивает данные на кластеры путем разбиения их на иерархичные группы, т.е. получаемое множество кластеров имеет иерархическую структуру. Принцип работы иерархической агломеративной процедуры состоит в последовательном объединении групп элементов, сначала самых близких, а затем всё более отдалённых друг от друга.<br>
**Скорость работы Single Link алгоритма - O(n^2), где n - количество элементов.**

Расстояние между одноэлементными кластерами определяется через расстояние между объектами: $\mathrm{R}(\{x\}, \{y\}) = \rho(x, y)$. Для определения расстояния между объектами вычисляются через:

**Евклидово расстояние**
$\rho (x, x')=\sqrt{\sum_{i}^{n}(x_{i}-x'_{i})^{2}}$

Есть ещё "Квадрат евклидова расстояния", "Расстояние городских кварталов (манхэттенское расстояние)", "Расстояние Чебышева"

Для вычисления расстояния $\mathrm{R}(U, V)$ между кластерами $\mathrm{U}$ и $\mathrm{V}$ на практике используются различные функции в зависимости от специфики задачи. Но мы расмотрим функцию расстояния single link

**Метод одиночной связи** (англ. single linkage)
<br><br>
$\mathrm{R_{min}}(U, V) = \displaystyle\min_{u \in U, v \in V} \rho(u, v)$

Для того, чтобы на каждом шаге быстро подсчитывать расстояние от образовавшего кластера $\mathrm{W}=\mathrm{U}\cup\mathrm{V}$ до любого другого кластера $\mathrm{S}$, используя известные расстояния с предыдущих шагов, для этого используют формулу Ланса-Уильямса:

$$\mathrm{R}(W, S) = \alpha_U \cdot \mathrm{R}(U, S) + \alpha_V \cdot \mathrm{R}(V, S) + \beta \cdot \mathrm{R}(U, V) + \gamma \cdot |\mathrm{R}(U, S) - \mathrm{R}(V, S)|$$, где $\alpha_U, \alpha_V, \beta, \gamma$ - числовые параметры.

В методе **одиночной связи(single link)** функция удовлетворяет формулу Ланса-Уильямса со следующим коэфициентом:
$\alpha_U = \dfrac{1}{2}, \alpha_V = \dfrac{1}{2}, \beta = 0, \gamma = -\dfrac{1}{2}$

### Свойство монотонности

Введем обозначение $\mathrm{R_t}$ - расстояние между кластерами, выбранными на шаге $t$ для объединения.

Дендрограмма позволяет представлять зависимости между множеством объектов с любым числом заданных характеристик
на двумерном графике, где по одной из осей откладываются все объекты, а по другой - расстояние $\mathrm{R_t}$.
Если не накладывать на это расстояние никаких ограничений, то дендрограмма будет иметь большое число самопересечений и изображение перестанет быть наглядным.
Чтобы любой кластер мог быть представлен в виде непрерывного отрезка на оси объектов и ребра не пересекались,
необходимо наложить ограничение монотонности на $\mathrm{R_t}$.

**Определение**<br>
Функция расстояния $\mathrm{R}$ является __монотонной__, если на каждом следующем шаге расстояние между кластерами не уменьшается:
$\mathrm{R_2} \leqslant \mathrm{R_3} \leqslant \dots \leqslant \mathrm{R_m}$

Расстояние является монотонным, если для коэффициентов в формул Ланса-Уильямса верна теорема Миллигана.
Теорема Миллигана: <br>
Если выполняются следующие три условия, то кластеризация является монотонной:
1. $\alpha_U \geqslant 0, \alpha_V \geqslant 0 $;<br>
2. $\alpha_U + \alpha_V + \beta \geqslant 1 $;<br>
3. $\min\{\alpha_U, \alpha_V\} + \gamma \geqslant 0 $.

### Алгоритм

def HierarchyCluster $(X=\begin{Bmatrix}x_{1},...,x_{n}\end{Bmatrix})$
* Инициализируем $C=X, G=X$.
* Пока в C больше одного элемента:
    * Выбираем два элемента $C$ $c_{1}$ и $c_{2}$ между которыми минимально.
    * Добавляем в $G$ вершину $c_{12}$, соединяем ее с вершинами $c_{1}$ и $c_{2}$
    * $C:= C\cup \begin{Bmatrix}x_{12}\end{Bmatrix}\setminus\begin{Bmatrix}c_{1}, c_{2}\end{Bmatrix}$
* Выдаем G.

### Реализация на Python

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from matplotlib import pyplot as plt
%matplotlib inline

# библиотеку scipy для иерархической кластеризации
from scipy.cluster.hierarchy import dendrogram, linkage  

# библиотека sklearn для иерархической кластеризации
from sklearn.cluster import AgglomerativeClustering
```

Создаем исскуственные данные

```python
X = np.array([[23,24],  
    [21,25],
    [24,16],
    [22,27],
    [42,28],
    [14,29],
    [25,23],
    [42,32],
    [42,34],
    [41,36],])
```

```python
labels = range(1, 11)  
plt.figure(figsize=(10, 7))  
plt.subplots_adjust(bottom=0.1)  
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):  
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()  
```

```python
linked = linkage(X, 'single')

labelList = range(1, 11)

plt.figure(figsize=(10, 7))  
dendrogram(linked,  
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

В этом примере мы видим что у нас появилось два кластера разделённые разными цветами, всё зависит от того, где мы разделим нашу дендрограмму.

### Попробуем подпортить данные и построить дендрограмму

```python
B = np.array([[9,14],  
    [10,15],
    [11,16],
    [12,17],
    [13,18],
    [14,19],
    [15,20],
    [17,22],
    [19,24],
    [21,26],])
```

```python
labels = range(1, 11)  
plt.figure(figsize=(10, 7))  
plt.subplots_adjust(bottom=0.1)  
plt.scatter(B[:,0],B[:,1], label='True Position')

for label, x, y in zip(labels, B[:, 0], B[:, 1]):  
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()  
```

```python
linked = linkage(B, 'single')

labelList = range(1, 11)

plt.figure(figsize=(10, 7))  
dendrogram(linked,  
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

В этом примере, мы намеренно подпортили данные, тут видно что дендрограмма стала не очень информативной, становится не понятно на сколько кластеров стоит делить дендрограмму.

### Кластеризация датасета **Iris**

```python
fig = plt.figure(figsize=(15, 30))
fig.patch.set_facecolor('white')

iris = datasets.load_iris()

mergings = linkage(iris.data, method='single')
R = dendrogram(mergings, labels=[iris.target_names[i] for i in iris.target], orientation = 'left', leaf_font_size = 12)

plt.show()
```

Тут я применил кластеризацию на данных **Iris**. Видим что у нас появляется два кластера, хотя судя по названиям классов, тут больше кластеров. Класс Setosa был хорошо класифицирован. Для того чтобы лучше класифицировать на этих данных класс цветов, есть другие функции расстояния, которые лучше класифицируют эти данные.

### Делаем кластеризацию с помощью библиотеки Sklearn

В параметрах указываем affinity=euclidian, linkage = single **(single link)**

```python
cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', compute_full_tree='auto', linkage='single')  
cluster.fit_predict(iris.data)
```

```python
plt.scatter(iris.data[:,0],iris.data[:,1], c=cluster.labels_, cmap='rainbow');
```

На этом графие видно, что кластеризация сделана тем же методом, что и ранее, но использована другая библиотека. В целом, тут тоже получается примерно - 2 клааса. Хотя в параметрах мы задавали 4 кластера, поэтому на графике 4 цвета.
