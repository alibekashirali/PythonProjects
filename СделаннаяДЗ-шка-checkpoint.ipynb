```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('train.csv')
data.head()
```

```python
data.info()
```

```python
data.drop(labels=['PoolQC', 'Id'], axis=1, inplace=True)
for i in data:
  if (data[i].dtype == 'float' or data[i].dtype == 'int'):
    data[data[i].isna()] = data[i].median()
data = pd.get_dummies(data, dummy_na=True)  #dummy_na=True - NaN-ы выделяются в отдельные столбцы
data.info(1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
scaled_data = pd.DataFrame(scaled_data, columns=data.columns)
#Стандартизация от 0 до 1
#вычитаем минимальное, прижимая к оси, делим на среднее, чтобы достич промежутка от минус половины до плюс половины
scaled_data.shape
```

```python
scaled_data.describe()
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split
model = LinearRegression()
X = scaled_data.drop(labels=['SalePrice'], axis=1)
y = scaled_data['SalePrice']
trainX, testX, trainY, testY = train_test_split(X, y, test_size=0.2)
model.fit(trainX, trainY)
#разбиваем трейн-файл на трейн и тест
```

```python
np.set_printoptions(suppress=True)
from sklearn.metrics import mean_squared_log_error, mean_squared_error
predictedY = model.predict(testX)
#Вполне может поругаться на то, что получились при расчете логарифмической MSE неадекватные величины
#print('log MSE: '+str(mean_squared_log_error(testY, predictedY))) - временно, т.к. почему-то невсегда работает
print('MSE: '+str(mean_squared_error(testY, predictedY)))
```

```python
test_data = pd.read_csv('test.csv')
test_data.info()
```

```python
test_data.drop(labels=['PoolQC', 'Id'], axis=1, inplace=True)

for i in test_data:
  if (test_data[i].dtype == 'float' or test_data[i].dtype == 'int'):
    test_data[test_data[i].isna()] = test_data[i].median()
    
test_data = pd.get_dummies(test_data, dummy_na=True)

test_scaled_data = scaler.fit_transform(test_data)
test_scaled_data = pd.DataFrame(test_scaled_data)

test_scaled_data.shape
# ВНЕЗАПНО количество столбцов не совпадает с количеством из train-выборки. Почему бы?
# Нельзя применять модель, т.к. она обучена на другом количестве признаков
# Требуется хэширование значений для категориальных признаков - преобразование из простого в сложное (причем нет обратного преобразования)
```

```python
#Вернемся к train-выборке. "План не удался" - Никита
#Линейная регрессия с L1-регуляризацией
lasso = Lasso(alpha=0.01)
lasso.fit(trainX, trainY)
predictedY = lasso.predict(testX)
print('log MSE: '+str(mean_squared_log_error(testY, predictedY)))
print('MSE: '+str(mean_squared_error(testY, predictedY)))
```

```python
#Линейная регрессия с L2-регуляризацией
ridge = Ridge(alpha=0.01)
ridge.fit(trainX, trainY)
predictedY = ridge.predict(testX)
print('log MSE: '+str(mean_squared_log_error(testY, predictedY)))
print('MSE: '+str(mean_squared_error(testY, predictedY)))
```

```python
%%time

from sklearn.preprocessing import PolynomialFeatures

# степень полинома
degree = 2

poly = PolynomialFeatures(degree)
polyTrainX = poly.fit_transform(trainX)
polyTestX = poly.fit_transform(testX)
ridge = Lasso(alpha=0.01)
ridge.fit(polyTrainX, trainY)
predictedY = ridge.predict(polyTestX)
print('log MSE '+str(degree)+' degree: '+str(mean_squared_log_error(testY, predictedY)))
print('MSE '+str(degree)+' degree: '+str(mean_squared_error(testY, predictedY)))
  
# Видим, что при тех же условиях регуляризации показатели ухудшились - усилилось переобучение, 
# что логично, учитывая соотношение количества признаков и объектов в выборке (иначе говоря, нам не нужна настолько усложненная модель)
```
